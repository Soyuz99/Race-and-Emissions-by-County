Used `sdf2 = df2[[    'County',
    'State',
    'Ethnicities.American Indian and Alaska Native Alone',

]]` to select the data needed to join the two datasets by creating a subset of data with the selected columns.

```python
sdf2 = df2[[
    'County',
    'State',
    'Ethnicities.American Indian and Alaska Native Alone',
'Ethnicities.Asian Alone',
'Ethnicities.Black Alone',
'Ethnicities.Hispanic or Latino',
'Ethnicities.Native Hawaiian and Other Pacific Islander Alone',
'Ethnicities.Two or More Races',
'Ethnicities.White Alone',
'Ethnicities.White Alone\t not Hispanic or Latino',
]]
```

In order to `.join` two data sets, the columns title needs to be the same, so use `sdf2['County'] = sdf2['County'].str.upper()` to set the sdf2 "County" to all uppercase so it matches the sdf "County".

```python
sdf2['County'] = sdf2['County'].str.upper()
```

```python
sdf2.head()
```

Used `sdf['County'] = sdf['Reported County']` to create an overlapping title of the column for `.join`

```python
sdf['County'] = sdf['Reported County']
```

```python
sdf.head()
```

Used `sdf2.set_index('County', inplace=True)` to create an overlapping index to `.join` through.

```python
sdf2.set_index('County', inplace=True)
```

`.set_index` cannot be ran twice in the same code block, so use `sdf.set_index('County', inplace=True)` to create overlapping index. The `inplace=True` is used to change the variable directly rather than creating a copy and then changing it. Used `sdf.join(sdf2,lsuffix='_Emissions', how= 'inner')` to join the two data sets. `how= 'inner'` is used to determine how the data set will be joined, in this case it is joined through overlapping variables of both sets of dat.

```python
sdf.set_index('County', inplace=True)

fdf = sdf.join(sdf2,lsuffix='_Emissions', how= 'inner')
```
